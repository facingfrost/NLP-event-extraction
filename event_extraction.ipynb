```python
import pandas as pd
df = pd.DataFrame({"date":date_list,"title":title_list})
df.head(10)
```

```python
df['date']=pd.to_datetime(df['date'])
df.head(10)
```

```python
import spacy
import tqdm

nlp = spacy.load('en_core_web_lg')

sent_vecs = {}
docs = []

for i in range(len(title_list)):
    doc = nlp(title_list[i])
    docs.append(doc)
    sent_vecs.update({title_list[i]:doc.vector})
    
sentences = list(sent_vecs.keys())
vectors = list(sent_vecs.values())
```

```python
sentences[:10]
```

```python
vectors[0]
```

```python
from sklearn.cluster import DBSCAN
import numpy as np
x = np.array(vectors)
# n_classes = {}
# for i in np.arange(0.001,1,0.002):
#     dbscan = DBSCAN(eps=i,min_samples=2,metric='cosine').fit(x)
#     n_classes.update({i:len(pd.Series(dbscan.labels_).value_counts())})

dbscan = DBSCAN(eps=0.08,min_samples=2,metric='cosine').fit(x)
```

```python
results = pd.DataFrame({"label":dbscan.labels_,"sent":sentences})
example_result = results[results.label == 1].sent.to_list()
example_result
```

```python
event_df = df[df.title.isin(example_result)][['date','title']]
event_df = event_df.sort_values(by='date').dropna()
event_df
```

```python
from sklearn.metrics import pairwise_distances_argmin_min
def get_mean_vector(sents):
    a = np.zeros(300)
    for i in range(len(sents)):
        a = a+ nlp(sents[i]).vector
    return a/len(sents)

def get_central_vector(sents, mean_vec):
    vecs = []
    for sent in sents:
        doc = nlp(sent)
        vecs.append(doc.vector)
    index = pairwise_distances_argmin_min(np.array([mean_vec]),vecs)[0][0]
    return sents[index]
```

```python
mean_vector = get_mean_vector(list(event_df['title']))
```

```python
mean_vector
```

```python
central_vector = get_central_vector(list(event_df['title']),mean_vector)
central_vector
```

```python
central_vectors = []
```

```python
for i in range(1000):
    example_result = results[results.label == i].sent.to_list()
    if len(example_result) >= 3:
        event_df = df[df.title.isin(example_result)][['date','title']]
        event_df = event_df.sort_values(by='date').dropna()
        mean_vector = get_mean_vector(list(event_df['title']))
        central_vector = get_central_vector(list(event_df['title']),mean_vector)
        central_vectors.append(central_vector)
```
